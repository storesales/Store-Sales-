##Data transformation

```python
# split order date column
df['Order Date'] = df['Order Date'].astype(str)
```

```python
df[['Day', 'Month', 'Year']] = df['Order Date'].str.split('/', expand=True)
```

```python
df.head()
```

```python
# make the date as pandas format 
df['Order Date Fixed'] = pd.to_datetime(
    df[['Month', 'Day', 'Year']].astype(str).agg('-'.join, axis=1),
    format="%m-%d-%Y"
)
```

```python
# Order Month column 
df['Order Month'] = df['Order Date Fixed'].dt.month
# Month Name 
df['Order Month'] = df['Order Date Fixed'].dt.strftime('%B')
```

```python
df['Order Year'] = df['Order Date Fixed'].dt.year
```

```python
# Drop columns 
# (day - month - year ) not needed anymore
df.drop(columns=['Day', 'Month', 'Year'], inplace=True)

```

```python
# Drop Order Date (Wrong Format)
df.drop(columns=['Order Date'], inplace=True)
```

```python
# Rename Order Date Fixed 
df.rename(columns={'Order Date Fixed': 'Order Date'}, inplace=True)
```

```python
# split ship date column
df['Ship Date'] = df['Ship Date'].astype(str)
df[['Ship_Day', 'Ship_Month', 'Ship_Year']] = df['Ship Date'].str.split('/', expand=True)
```

```python
# fixed ship date
df['Ship Date Fixed'] = pd.to_datetime(
    df[['Ship_Month', 'Ship_Day', 'Ship_Year']].astype(str).agg('-'.join, axis=1),
    format="%m-%d-%Y"
)
```

```python
# drop columns 
df.drop(columns=['Ship_Day', 'Ship_Month', 'Ship_Year', 'Ship Date'], inplace=True)
```

```python
# rename fixed ship date 
df.rename(columns={'Ship Date Fixed': 'Ship Date'}, inplace=True)
```

```python
# add delivery days column 
df['Delivery Days'] = (df['Ship Date'] - df['Order Date']).dt.days
```

```python
df["Postal Code"]=df["Postal Code"].astype(str)
```

```python
df["City"] = df["City"].astype("category")
df["State"] = df["State"].astype("category")
df["Country"] = df["Country"].astype("category")
df["Segment"] = df["Segment"].astype("category")
df["Ship Mode"] = df["Ship Mode"].astype("category")
```

```python
print(df.dtypes)
```

```python
df["Cost Per Unit"]=df["Cost"]/df["Quantity"]
```

```python
df["Selling Price Per Unit"]=df["Sales"]/df["Quantity"]
```

```python
df["Total Profit"]=(df["Sales"]-df["Cost"])*df["Quantity"]
```

```python
df.rename(columns={"Profit":"Profit Per Unit"}, inplace=True)
```

```python
df.head()
```

```python
# outliers
numeric_cols = df.select_dtypes(include='number').columns
cols_to_check = [col for col in numeric_cols if col not in ['Row ID', 'Postal Code', 'Order Year']]

for col in cols_to_check:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    outliers = df[(df[col] < (Q1 - 1.5*IQR)) | (df[col] > (Q3 + 1.5*IQR))]
    print(f"{col}: {len(outliers)} outliers")
```

```python
import matplotlib.pyplot as plt
numeric_cols = df.select_dtypes(include='number').columns
cols_to_plot = [col for col in numeric_cols if col not in ['Row ID', 'Postal Code', 'Order Year']]
for col in cols_to_plot:
    plt.figure(figsize=(6,4))
    plt.boxplot(df[col].dropna())
    plt.title(f'Boxplot of {col}')
    plt.show()
```
